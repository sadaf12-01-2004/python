```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a dataframe called 'df'

# Filter data into recession and non-recession
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Combine the two datasets for comparison
combined_data = pd.concat([recession_data.assign(Period="Recession"),
                           non_recession_data.assign(Period="Non-Recession")])

# Create the line plot to compare sales trends per vehicle type
plt.figure(figsize=(12, 6))
sns.lineplot(data=combined_data, x='Date', y='Automobile_Sales', hue='Vehicle_Type', 
             style='Period', markers=True)

# Add labels and title
plt.title("Comparison of Automobile Sales Trends: Recession vs Non-Recession Periods")
plt.xlabel('Date')
plt.ylabel('Automobile Sales')

# Show the plot
plt.xticks(rotation=45)
plt.legend(title="Vehicle Type and Period")
plt.tight_layout()
plt.show()
```
